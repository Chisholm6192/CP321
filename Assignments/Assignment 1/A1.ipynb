```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('energy-price.csv')
df.head(10)
```

## Question 1

Compare the mean prices of Regular Unleaded Gasoline among different cities ('Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor','London', 'Peterborough', 'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay', 'Timmins', 'Kenora', 'Parry Sound',"St. Catharine's") in 2023, and plot a corresponding bar chart. Please ensure that the bars are sorted by the price amounts.

```python
Data = df[(df['Fuel Type']=="Regular Unleaded Gasoline") & (df['Date'] >= "2023-01-01") & (df['Date'] <= "2023-12-31")]

#filtering and dropping columns
Data = Data.drop('_id', axis=1)
Data = Data.drop('Date', axis=1)
Data = Data.drop('Type de carburant', axis=1)
Data = Data.drop('Fuel Type', axis=1)
Data = Data.drop("Northern Average/Moyenne du nord de l'Ontario", axis=1)
Data = Data.drop("Southern Average/Moyenne du sud de l'Ontario", axis=1)
Data = Data.drop("Ontario Average/Moyenne provinciale", axis=1)
'''The "Data" dataframe now represents the raw data of the CSV file, filtered to only show data from 2023,
and of regular unleadead gasoline. Dropping the columns which don't represent data that contains values for a city'''


#computing and rearranging dataframe
mean = pd.DataFrame(df)
mean_prices = Data.mean()
mean_df = pd.DataFrame({"Mean Gas Price": mean_prices})
mean_df = mean_df.sort_values(by='Mean Gas Price') #sorting values in ascending order

plt.bar(mean_df.index, mean_df['Mean Gas Price']) #plot data

#styling of plot
plt.title("Average Prices of Regular Unleaded Gasoline Across Ontario in 2023")
plt.xticks(rotation="vertical")
plt.ylim(120, 180) #zooming in on data to see the differences better
plt.xlabel("City")
plt.ylabel("Avg Gas Prices ($ CAD)")

plt.show()
```

## Question 4

Plot the normalized ECDF (Empirical Cumulative Distribution Function) of the Regular Unleaded Gasoline prices in Toronto West/Ouest for 2023.

```python
#uses the Toronto West Dataframe from Question 3

#ECDF Function
def ecdf(data):
    length_data = len(data)
    x = np.sort(data)
    y = np.arange(1, length_data+1) / length_data
    return x, y

x, y = ecdf(TorW_df)

plt.plot(x, y) #plot data

#styling of plots
plt.title("Distribution of Regular Unleaded Gasoline Prices in West Toronto in 2023")
plt.xlabel("Gas Prices ($ CAD)")
plt.ylabel("ECDF")

plt.show()
```
